```python

# %%
import faiss
import pandas as pd
import numpy as np

# %% [markdown]
# ### 🪷 faiss metric으로 output token과 emoji의 유사성 계산

# %%
# from google.colab import drive
# drive.mount('/content/drive/')

vectors = []
df = pd.read_csv('emoji_vector_unicode직접기입.csv')


# %%
import torch

vector_set = np.array(vector_set)
vector_set = vector_set.astype('float32')
print(vector_set.shape)

# change vector type from numpy array to torch tensor
# vector_set = torch.tensor(vector_set)

# %%
#vector_array = vector_set.astype(np.float32)

# %% [markdown]
# ---

# %%
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("augustinLib/text-emoji-encoder")
model = AutoModelForSequenceClassification.from_pretrained("augustinLib/text-emoji-encoder")

# %%
import torch

# %% [markdown]
# 이 시점에서 data_list 불러오기

# %%
input_sequence = save_list
print(input_sequence)
tokenized_input = tokenizer(input_sequence, return_tensors="pt", padding='max_length')
logits = model(input_ids = tokenized_input.input_ids,
               attention_mask = tokenized_input.attention_mask).logits

query_vector = logits.detach().numpy()

# %%
dimension = 300
faiss_index = faiss.IndexFlatIP(dimension)
faiss_index = faiss.IndexIDMap(faiss_index)
print(faiss_index.ntotal)

faiss_index.add_with_ids(vector_set, np.arange(len(vector_set)))
```

```python
# %%
Distance, Index = faiss_index.search(query_vector, 10)

# %%
print(f'Index:{Index}')
print(f'Distance:{Distance}')

# %%
df.loc[Index[0], :]
```
